```python
import pandas as pd
```

```python
import requests

nhn_url = 'https://finance.naver.com/sise/field_submit.nhn' \
    '?menu=market_sum' \
    '&returnUrl=http%3A%2F%2Ffinance.naver.com%2Fsise%2Fsise_market_sum.nhn?&page={}' \
    '&fieldIds=market_sum' \
    '&fieldIds=operating_profit' \
    '&fieldIds=open_val' \
    '&fieldIds=property_total' \
    '&fieldIds=debt_total' \
    '&fieldIds=listed_stock_cnt'

res = requests.get(nhn_url.format(1))
```

```python
df3 = pd.read_html(res.text)
```

```python

df4 = df3[1][ df3[1]['N'].notna()]
df4.head()
```

```python
df5 = df4.loc[:, ['N', '종목명', '현재가', '시가', '상장주식수', '시가총액', '자산총계', '부채총계', '영업이익']]
df5['N'] = df5['N'].astype('int')
df5.index = df5['N']
df5 = df5.drop(axis=1, columns='N')
df5.tail()
```

```python
naverfin = df5

num_total = 32


encoded_url = 'https://finance.naver.com/sise/field_submit.nhn?menu=market_sum&returnUrl=http%3A%2F%2Ffinance.naver.com%2Fsise%2Fsise_market_sum.nhn%3F%26page%3D{}&fieldIds=market_sum&fieldIds=operating_profit&fieldIds=open_val&fieldIds=property_total&fieldIds=debt_total&fieldIds=listed_stock_cnt'


for i in range(2, num_total+1):
    
    print(nhn_url.format(i))
    res = requests.get(encoded_url.format(i))
    
    tmp_df = pd.read_html(res.text)
    #print(tmp_df[1].columns)

    tmp_df = tmp_df[1][tmp_df[1]['N'].notna()].loc[:, ['N', '종목명', '현재가', '시가', '상장주식수', '시가총액', '자산총계', '부채총계', '영업이익']]
    
    
    tmp_df['N'] = tmp_df['N'].astype('int')
    tmp_df.index = tmp_df['N']
    tmp_df = tmp_df.drop(axis=1, columns='N')
    print(tmp_df.head())
    
    naverfin = naverfin.append(tmp_df)
    
naverfin.shape
    
    
    
    
```

```python
naverfin['myRoe'] = naverfin['영업이익']/(naverfin['자산총계']-naverfin['부채총계'])
```
