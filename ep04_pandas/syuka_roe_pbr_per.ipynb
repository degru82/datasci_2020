```python
import pandas as pd
```

```python
import requests

nhn_url = 'https://finance.naver.com/sise/field_submit.nhn' \
    '?menu=market_sum' \
    '&returnUrl=http%3A%2F%2Ffinance.naver.com%2Fsise%2Fsise_market_sum.nhn?&page={}' \
    '&fieldIds=market_sum' \
    '&fieldIds=operating_profit' \
    '&fieldIds=open_val' \
    '&fieldIds=property_total' \
    '&fieldIds=debt_total' \
    '&fieldIds=listed_stock_cnt'

res = requests.get(nhn_url.format(1), verify=False)
```

```python
df3 = pd.read_html(res.text)
df3[1]
```

```python
df4 = df3[1][ df3[1]['N'].notna()]
df4.head()
```

```python
df5 = df4.loc[:, ['N', '종목명', '현재가', '시가', '상장주식수', '시가총액', '자산총계', '부채총계', '영업이익']]
df5['N'] = df5['N'].astype('int')
df5.index = df5['N']
df5 = df5.drop(axis=1, columns='N')
df5.columns, df5.index
```

```python
naverfin = pd.DataFrame()
num_total = 32

query_url = 'https://finance.naver.com/sise/field_submit.nhn?menu=market_sum&returnUrl=http%3A%2F%2Ffinance.naver.com%2Fsise%2Fsise_market_sum.nhn%3F%26page%3D{}&fieldIds=market_sum&fieldIds=operating_profit&fieldIds=open_val&fieldIds=property_total&fieldIds=debt_total&fieldIds=listed_stock_cnt'

for i in range(1, num_total+1):
    
    #print(nhn_url.format(i))
    res = requests.get(query_url.format(i), verify=False)
    
    tmp_df = pd.read_html(res.text)
    #print(tmp_df[1].columns)

    tmp_df = tmp_df[1][tmp_df[1]['N'].notna()].loc[:, ['N', '종목명', '현재가', '상장주식수', '시가총액', '자산총계', '부채총계', '영업이익']]
    
    
    tmp_df['N'] = tmp_df['N'].astype('int')
    tmp_df.index = tmp_df['N']
    tmp_df = tmp_df.drop(axis=1, columns='N')
    print(tmp_df.head())
    
    naverfin = naverfin.append(tmp_df)
    
naverfin.shape
    
    
    
    
```

```python
naverfin
```

```python
naverfin[81:100]
```

```python
naverfin.shape
```

```python
naverfin['myRoe'] = naverfin['영업이익']/(naverfin['자산총계']-naverfin['부채총계'])
```

```python
naverfin[80:100]
```

```python
top_roe = naverfin.sort_values(by=['myRoe'], ascending=False)
```

```python
top_roe
```

__아니 이럴수가,  
영업이익이 음수인데도 ROE가 높다는게 말이되나?  
하고 봤더니, 부채가 자산을 눌렀네

```python
naverfin['자본'] = naverfin['자산총계'] - naverfin['부채총계']
naverfin.head()
```

```python
naverfin.head()
```

```python
naverfin.sort_values(by=['자본', 'myRoe'], ascending=[False, False])
```

# 친절한 이웃, 네이버에서 ROE, PBR, PER까지 가져오자
- 사실은 그냥 처음부터 데이터프레임 만들때 가져왔으면 되었지만,
- 연습도 할 겸, 두 개의 데이터 프레임을 옆으로 붙이는 걸 해보겠습니다

```python
kindnaver_url = 'https://finance.naver.com/sise/field_submit.nhn?menu=market_sum&returnUrl=http%3A%2F%2Ffinance.naver.com%2Fsise%2Fsise_market_sum.nhn%3F%26page%3D{}&fieldIds=per&fieldIds=roe&fieldIds=pbr'

res = requests.get(kindnaver_url.format(1), verify=False)
tmp_df = pd.read_html(res.text)

tmp_df[1].head()
```

```python
res = requests.get
```

```python
indicators = pd.DataFrame()

for i in range(1, num_total+1):
    res = requests.get(kindnaver_url.format(i), verify=False)
    
    tmp_df = pd.read_html(res.text)
    #print(tmp_df[1].columns)

    tmp_df2 = tmp_df[1][tmp_df[1]['N'].notna()].loc[:, ['N', '종목명', 'PER', 'ROE', 'PBR']]
    
    
    tmp_df2['N'] = tmp_df2['N'].astype('int')
    tmp_df2.index = tmp_df2['N']
    tmp_df2 = tmp_df2.drop(axis=1, columns='N')
    print(tmp_df2.head())
    
    indicators = indicators.append(tmp_df2)
    
indicators.shape
    
```

```python
naverfin.shape, indicators.shape
```

```python
indicators[80:100]
```

```python
fullsize = pd.concat([naverfin, indicators.loc[:, ['PER', 'ROE', 'PBR']]], axis=1, sort=False)
fullsize
```

# 투자조건 검토
- PER, PBR 높은거 보자
