```python
import numpy as np
```

```python
def rmsle(y, y_, convertExp=True):
    """
    Root Mean Square Logarithmic Error
    """
    
    if convertExp:
        y = np.exp(y),
        y_ = np.exp(y_)
    log1 = np.nan_to_num(np.array([np.log(v + 1) for v in y]))
    log2 = np.nan_to_num(np.array([np.log(v + 1) for v in y_]))
    calc = (log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
```

# LINEAR MODEL

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
import warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings("ignore", category=DeprecationWarning)

# Initialize logistic regression model
lModel = LinearRegression()

# Train the model
y_labellog = np.log1p(y_train)
lModel.fit(X = df_train,y = y_labellog)

# Make predictions
preds = lModel.predict(X=df_train)
print ("RMSLE Value For Linear Regression: ",rmsle(np.exp(preds), np.exp(y_labellog), False))
```

# ENSEMBLE MODEL - RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
rfModel = RandomForestRegressor(n_estimators=10)
yLabelsLog = np.log1p(y_train)
rfModel.fit(df_train,yLabelsLog)
preds = rfModel.predict(X= df_train)
print ("RMSLE Value For Random Forest: ",rmsle(np.exp(yLabelsLog),np.exp(preds),False))
```
